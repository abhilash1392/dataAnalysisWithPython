```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv('johnsonFiltration.csv')
df
```

```python
just_dummies=pd.get_dummies(df['type_of_repair'])
just_dummies
```

```python
df=pd.concat([df,just_dummies],axis=1)
df
```

```python
df.drop(['type_of_repair','m'],axis=1,inplace=True)
```

```python
X=df.drop(['service_call','repair_time_in_hours'],axis=1)
Y=df.repair_time_in_hours
```

```python
import statsmodels.api as sm
```

```python
X=sm.add_constant(X)
reg=sm.OLS(Y,X).fit()
reg.summary()
```

```python
from sklearn.linear_model import LinearRegression
X=df.drop(['service_call','repair_time_in_hours'],axis=1).values
Y=df.repair_time_in_hours.values
reg=LinearRegression().fit(X,Y)
reg.score(X,Y)
```
