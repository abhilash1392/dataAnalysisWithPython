```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from statsmodels.formula.api import ols 
import statsmodels.api as sm 
```

```python
df1=pd.read_csv('trucking.csv')
df1
```

```python
# Using only one variable 
reg1=ols(formula='travel_time ~ x1',data=df1)
fit1=reg1.fit()
fit1.summary()
```

```python
anova_table=sm.stats.anova_lm(fit1)
print(anova_table)
```

```python
# Using both variables
reg2=ols(formula='travel_time ~ x1+n_of_deliveries',data=df1)
fit2=reg2.fit()
fit2.summary()
```

```python
anova_table=sm.stats.anova_lm(fit2)
print(anova_table)
```
